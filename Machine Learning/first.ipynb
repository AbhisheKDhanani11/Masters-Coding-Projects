```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
print(df.isnull().sum().sum())
df
```

```python
df['age2'] = df['age'].replace(("?"), (np.nan))
df[['name', 'age', 'age2']].sample(100)
```

```python
df2 = df[df['age2'].isnull()==False]
df2
```

```python
def get_province(dest):
    if dest and ',' in dest:
        return dest.split(',')[1].strip()
    else:
        return None
    
get_province(None)
get_province('?')
df['province'] = df['home.dest'].apply(get_province)
df['province'].value_counts()
```

```python
def get_Fname(name):
    if name and ',' in name:
        return name.split(',')[0].strip()
    else:
        None
        
df['FName'] = df['name'].apply(get_Fname)
df['FName'].value_counts()
```
