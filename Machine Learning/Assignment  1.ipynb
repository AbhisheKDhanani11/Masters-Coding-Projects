```python
# using to numeric to convert all the values of fare columns to numeric or int values, error argument will fill NA if a value is not numeric 
df['fare'] = pd.to_numeric(df['fare'], errors='coerce')
df.groupby(['pclass'])['fare'].agg(['mean', 'max', 'min'])
```
