```python
import os

"""
Loading the data getting each line from the file, spliting the line into token which is first word, its position, chunk and NER tag 
and appending it into my sentence list and appending the each sentance into sentences if the next line is empty or start of doc and returing the sentances list.
"""
def load_data(file_path):
    sentences = []
    sentence = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if line.startswith('-DOCSTART-') or not line:
                if sentence:
                    sentences.append(sentence)
                    sentence = []
            elif line:
                token, pos, chunk, ner = line.split()
                sentence.append((token, ner))  
    if sentence:
        sentences.append(sentence)
    return sentences

# applying the loading fuction to all data 
train_data = load_data('D:/Masters/Sem 3/Text analysis/Data/NER/train.txt')
test_data = load_data('D:/Masters/Sem 3/Text analysis/Data/NER/test.txt')
valid_data = load_data('D:/Masters/Sem 3/Text analysis/Data/NER/valid.txt')

print("First train sentence:", train_data[0])
```

```python
# feature extraction 

# function to see if the word is capital or not
def is_capitalized(word):
    return word[0].isupper()

"""
This functions will get each sentance int my data and extract feature for model. it will take the sentance and its index in data as input.
Extract each words its position in sentance, chake if it is capital or not, add the previous word and next word and also start and end tag with previous and next tags.
"""
def extract_features(sentence, index):
    word, ner_tag = sentence[index]
    features = {
        'word': word,
        'pos': sentence[index][1].split('-')[0] if '-' in sentence[index][1] else 'O',  
        'capitalized': is_capitalized(word),
        'word_prev': sentence[index-1][0] if index > 0 else '<START>',
        'word_next': sentence[index+1][0] if index < len(sentence)-1 else '<END>',
        'tag_prev': sentence[index-1][1] if index > 0 else '<START>',
        'tag_next': sentence[index+1][1] if index < len(sentence)-1 else '<END>'
    }
    return features, ner_tag

# using first sentance in my data to see how this function work
first_sentence = train_data[0]
features_and_tags = [extract_features(first_sentence, i) for i in range(len(first_sentence))]
print("Features and tags for first sentence:", features_and_tags)
```

```python
import numpy as np
from gensim.models import Word2Vec
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Extract features for all sentences 
def prepare_features_and_tags(data):
    X, y = [], []
    for sentence in data:
        features_and_tags = [extract_features(sentence, i) for i in range(len(sentence))]
        for features, tag in features_and_tags:
            X.append([features['word'], features['pos'], features['capitalized'], 
                      features['word_prev'], features['word_next'], 
                      features['tag_prev'], features['tag_next']])
            y.append(tag)
    return X, y

X_train, y_train = prepare_features_and_tags(train_data)
X_test, y_test = prepare_features_and_tags(test_data)
X_valid, y_valid = prepare_features_and_tags(valid_data)

# Prepare sentences for Word2Vec training
all_sentences = [[token for token, _ in sentence] for sentence in train_data + test_data + valid_data]

# Train Word2Vec embeddings on your dataset
word2vec_model = Word2Vec(sentences=all_sentences, vector_size=50, window=5, min_count=1, workers=4)

# Create embedding lookup function
def get_embedding(word, model, size=50):
    try:
        return model.wv[word]
    except KeyError:
        return np.zeros(size)  # Return zero vector for unknown words

# Encode POS tags and context tags
pos_encoder = LabelEncoder()
tag_encoder = LabelEncoder()

# getting all the position tags and NER tags 
all_pos = [item[1] for item in X_train]
all_tags = y_train + y_test + y_valid  # NER tags
all_context_tags = [item[5] for item in X_train] + [item[6] for item in X_train]  # prev and next tags
all_tags.extend(['<START>', '<END>'])
all_tags.extend(all_context_tags)

# Encoding the tags
pos_encoder.fit(all_pos)
tag_encoder.fit(all_tags)

# Function to encode features with embeddings
def encode_features_with_embeddings(X, embedding_size=50):
    encoded = []
    for x in X:
        word_embedding = get_embedding(x[0], word2vec_model, embedding_size)  # Current word embedding
        word_prev_embedding = get_embedding(x[3], word2vec_model, embedding_size)  # Previous word embedding
        word_next_embedding = get_embedding(x[4], word2vec_model, embedding_size)  # Next word embedding
        encoded_x = np.concatenate([
            word_embedding,  # 50-dimensional embedding
            [pos_encoder.transform([x[1]])[0]],  # POS tag
            [float(x[2])],  # Capitalized (True/False as float)
            word_prev_embedding,  # 50-dimensional embedding
            word_next_embedding,  # 50-dimensional embedding
            [tag_encoder.transform([x[5]])[0]],  # Previous tag
            [tag_encoder.transform([x[6]])[0]]  # Next tag
        ])
        encoded.append(encoded_x)
    return np.array(encoded)

# Encode all datasets with embeddings
embedding_size = 50
X_train_encoded = encode_features_with_embeddings(X_train, embedding_size)
X_test_encoded = encode_features_with_embeddings(X_test, embedding_size)
X_valid_encoded = encode_features_with_embeddings(X_valid, embedding_size)

# Encode labels
y_train_encoded = tag_encoder.transform(y_train)
y_test_encoded = tag_encoder.transform(y_test)
y_valid_encoded = tag_encoder.transform(y_valid)

# Convert labels to one-hot encoding
num_classes = len(tag_encoder.classes_)
y_train_cat = to_categorical(y_train_encoded, num_classes)
y_test_cat = to_categorical(y_test_encoded, num_classes)
y_valid_cat = to_categorical(y_valid_encoded, num_classes)

input_size = embedding_size * 3 + 4  # 50 (word) + 50 (word_prev) + 50 (word_next) + 1 (pos) + 1 (capitalized) + 1 (tag_prev) + 1 (tag_next)
print("X_train_encoded shape:", X_train_encoded.shape)
print("y_train_cat shape:", y_train_cat.shape)
print("Input size:", input_size)
print("Number of classes:", num_classes)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.optimizers import Adam

# Reshape data for LSTM (samples, timesteps, features) - here, each token is a timestep
X_train_reshaped = X_train_encoded.reshape((X_train_encoded.shape[0], 1, X_train_encoded.shape[1]))
X_valid_reshaped = X_valid_encoded.reshape((X_valid_encoded.shape[0], 1, X_valid_encoded.shape[1]))
X_test_reshaped = X_test_encoded.reshape((X_test_encoded.shape[0], 1, X_test_encoded.shape[1]))

# Build the LSTM model
model = Sequential()
model.add(LSTM(154, return_sequences=False, input_shape=(1, 154)))
model.add(Dense(77, activation='relu'))
model.add(Dense(39, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(X_train_reshaped, y_train_cat, 
                    validation_data=(X_valid_reshaped, y_valid_cat), 
                    epochs=10, batch_size=32, verbose=1)

# Evaluate on test set
loss, accuracy = model.evaluate(X_test_reshaped, y_test_cat, verbose=0)
print(f"Test accuracy: {accuracy:.4f}")
```

```python
# used internet for model evolution 
from sklearn.metrics import precision_recall_fscore_support
import numpy as np

# Get predictions for the test set using reshaped data
y_pred = model.predict(X_test_reshaped, verbose=0)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = y_test_encoded  # Already encoded

# Convert back to labels
y_pred_labels = tag_encoder.inverse_transform(y_pred_classes)
y_test_labels = tag_encoder.inverse_transform(y_test_classes)

# Function to extract entities from BIO tags (Beginning, inside, outside) for a sentence
def extract_entities(tags):
    entities = []
    current_entity = []
    for tag in tags:
        if tag.startswith('B-'):
            if current_entity:
                entities.append(' '.join(current_entity))
            current_entity = [tag.split('-')[1]]
        elif tag.startswith('I-') and current_entity:
            current_entity.append(tag.split('-')[1])
        elif tag == 'O' and current_entity:
            entities.append(' '.join(current_entity))
            current_entity = []
    if current_entity:
        entities.append(' '.join(current_entity))
    return entities if entities else ['O']  # Return ['O'] if no entities

# Process data sentence by sentence to align true and predicted entities
test_data_sentences = [sentence for sentence in test_data]  # Original sentences
true_entities_list = []
pred_entities_list = []

# Extract entities for each sentence with length checking
for sentence in test_data_sentences:
    sentence_length = len(sentence)
    start_idx = sum(len(s) for s in test_data_sentences[:test_data_sentences.index(sentence)])
    end_idx = start_idx + sentence_length
    true_tags = y_test_labels[start_idx:end_idx]
    pred_tags = y_pred_labels[start_idx:end_idx]
    
    if len(true_tags) != len(pred_tags):
        print(f"Warning: Mismatch in sentence length at index {test_data_sentences.index(sentence)}: "
              f"true={len(true_tags)}, pred={len(pred_tags)}")
        pred_tags = pred_tags[:len(true_tags)]  # Truncate to match true tags
    
    true_entities = extract_entities(true_tags)
    pred_entities = extract_entities(pred_tags)
    true_entities_list.extend(true_entities)
    pred_entities_list.extend(pred_entities)

# Ensure lists have the same length (pad with 'O' if necessary)
max_length = max(len(true_entities_list), len(pred_entities_list))
true_entities_list.extend(['O'] * (max_length - len(true_entities_list)))
pred_entities_list.extend(['O'] * (max_length - len(pred_entities_list)))

# Compute entity-level precision, recall, F1
precision, recall, f1, _ = precision_recall_fscore_support(true_entities_list, pred_entities_list, average='weighted', zero_division=0)
print(f"Entity-level Precision: {precision:.4f}")
print(f"Entity-level Recall: {recall:.4f}")
print(f"Entity-level F1-score: {f1:.4f}")
```
